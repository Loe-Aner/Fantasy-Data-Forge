ho(ho(ho())) # should return "Ho Ho Ho!"

```python
def ho(s: str | None = None):
    if s is None:
        return "Ho!"
    
    ho_count = s.count("Ho") + 1
    return ("Ho " * ho_count).rstrip() + "!"
```

```python
ho(ho(ho(ho())))
```

```python
# ====== NIE MOJE, ALE FAJNE ======
def ho(*args):
    if not args:
        return "Ho!"
    return "Ho " + args[0]
```

```python
ho(ho())
```
